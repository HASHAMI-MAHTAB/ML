Mobile Price Range

Objective : To find the price range of the Mobile.

Data Source : Data is collected from the kagle page of the YBI Foundation.

Import Library :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import Data :

```python
mobile = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/MobilePriceRange.csv")
```

Describe Data :

```python
mobile.head()
```

```python
mobile.tail()
```

Data Visualization :

```python
plt.figure(figsize = (19,6))
sns.heatmap(mobile.corr() , annot = True)
```

```python
plt. figure (figsize = (12, 6))
plt.subplot(1,2,1)
sns.barplot( x = 'PxHeight' , y = 'PxWeight' , data=mobile , palette = 'Blues')
```

Define Target Variable (y) and Feature Variables (X)

```python
y = mobile['PriceRange']
x = mobile.drop(['PriceRange'] , axis = 1)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

Modelling and Evaluation :

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=10)
model.fit(x_train , y_train)
```

```python
model.score(x_train , y_train)
```

Prediction :

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```
